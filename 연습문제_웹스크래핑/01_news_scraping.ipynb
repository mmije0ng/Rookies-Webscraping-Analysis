```python
import requests
from bs4 import BeautifulSoup

# 다음 경제 뉴스 URL
url = 'https://news.daum.net/economy'

# 요청
response = requests.get(url)
response.encoding = 'utf-8'

# 실행결과 출력 요구사항
print(url)
print(type(response))
print(response.status_code)
```

```python

# 응답이 정상일 경우
if response.status_code == 200:
    # BeautifulSoup 객체 생성
    soup = BeautifulSoup(response.text, 'html.parser')

    # 뉴스 리스트 선택
    li_tags = soup.select('ul.list_newsheadline2 li')
    print(type(li_tags), len(li_tags))

    # 뉴스 링크 + 제목 출력
    for li_tag in li_tags:
        a_tag = li_tag.find('a')
        strong_tag = li_tag.select_one('div.cont_thumb strong.tit_txt')

        if a_tag and strong_tag:
            link = a_tag['href']
            title = strong_tag.text.strip()

            print(link)
            print(title)

# 응답이 비정상일 경우
else:
    print(f'Error Code = {response.status_code}')
```

## 질문 2. 여러개의 section 중 하나를 선택해서 url에서 뉴스기사의 링크와 제목을 출력하는 코드를 함수로 작성하기

```python
import requests
from bs4 import BeautifulSoup
```

```python

section_dict = {
    '기후/환경': 'climate',
    '사회': 'society',
    '경제': 'economy',
    '정치': 'politics',
    '국제': 'world',
    '문화': 'culture',
    '생활': 'life',
    'IT/과학': 'tech',
    '인물': 'people'
}

def print_news(section_name):
    section = section_dict.get(section_name)

    if section is None:
        print('존재하지 않는 섹션입니다.')
        return

    url = f'https://news.daum.net/{section}'

    response = requests.get(url)
    response.encoding = 'utf-8'

    print(f'======> {url} {section_name} 뉴스 <======')

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')

        li_tags = soup.select('ul.list_newsheadline2 li')

        for li_tag in li_tags:
            a_tag = li_tag.find('a')
            strong_tag = li_tag.select_one('div.cont_thumb strong.tit_txt')

            if a_tag and strong_tag:
                link = a_tag['href']
                title = strong_tag.text.strip()

                print(link)
                print(title)
    else:
        print(f'Error Code = {response.status_code}')
```

```python
print_news('사회')
```

```python
print_news('정치')
```

```python
print_news('경제')
```
